### VALIDATING PERMUTATION POOLS AND CREATING CHAINS

```python
algo_list = [STKE]
full_chain = []
chain_len = 2
valid_comb = []
valid_set = []


while chain_len < 3:
    comb = itertools.combinations(asa_list,chain_len)
    for each in comb:
        strt = 0
        seq_len = len(each)
        # print('seq',each)
        while strt < seq_len-1:
            a = each[strt]
            b = each[strt+1]
            fp = client.fetch_pool(a,b)
            try:
                pool_dict = fp.info()
                # print('Passed', strt)            
                if strt == seq_len-2:
                    # print('yay',each)
                    full_chain = algo_list + list(set(each)) + algo_list
                    # full_chain = tuple(each)
                    v = full_chain
                    print(v)
                    valid_comb.append(v)
            
                
            except:
                break
            strt +=1
    chain_len +=1

with open('stke_comb.data', 'wb') as filehandle:
    # store the data as binary data stream
    pickle.dump(valid_comb, filehandle)
    print('Valid Chains Data Complete')
```
